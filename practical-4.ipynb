```python
########################################################################################################
#Name : - hiral shingadiya
#Enrollment :- 216270307061
#Subject :- IML
#Sem :- 5
########################################################################################################

#Practical - 4 : Write a NumPy program to implement following operation

#Practical - 4.3 to calculate mean across dimension, in a 2D numpy array

import numpy as np

arr = np.arange(1,10).reshape(3,3)
print("Array : \n",arr)

c = np.mean(arr)
print("\nColoum vise mean : ",c)
r = np.mean(arr,axis=0)
print("Row vise mean : ",r)
```
