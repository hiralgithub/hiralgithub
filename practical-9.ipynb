```python
########################################################################################################
#Name : - hiral shingadiya
#Enrollment :- 216270307061
#Subject :- IML
#Sem :- 5
########################################################################################################

#Practical - 9: Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted
#Closing prices and Volume of given company between two specific dates.


import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/ADANIPORTS.csv')

df['Date'] = pd.to_datetime(df['Date'])

start_date = '2007-11-27'  
end_date = '2008-02-06'   

mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
filtered_df = df.loc[mask]

plt.figure(figsize=(12, 6))

plt.plot(filtered_df['Date'], filtered_df['Open'], label='Open', linewidth=1)
plt.plot(filtered_df['Date'], filtered_df['High'], label='High', linewidth=1)
plt.plot(filtered_df['Date'], filtered_df['Low'], label='Low', linewidth=1)
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Close', linewidth=1)


plt.twinx()
plt.plot(filtered_df['Date'], filtered_df['Volume'], label='Volume', color='red', linewidth=1)

plt.title('Stock Price and Volume Between {} and {}'.format(start_date, end_date))
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```
